```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
url = "https://www-fars.nhtsa.dot.gov/Main/index.aspx"
r = requests.get(url)
```

```python
print(r)
```

```python
soup = bs(r.text)
table = soup.find('table', {'class': 'datatable01'})
columns = []
data = []

# Get the headers
headers = table.find_all('th')
for header in headers:
    print(header)
    columns.append(header.get_text())

# Get the rows
rows = table.find_all('tr')
for row in rows:
    cells = row.find_all('td')
    row_data = []
    for cell in cells:
        row_data.append(cell.get_text())
    if row_data:
        data.append(row_data)

#df = pd.DataFrame(data, columns=columns)
headers
```
