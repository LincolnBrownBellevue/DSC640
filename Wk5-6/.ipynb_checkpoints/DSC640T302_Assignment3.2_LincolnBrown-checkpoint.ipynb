```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import seaborn as sns
```

```python
## Import data
top_crimes_df = pd.read_excel('./ex4-2/crimerates-by-state-2005_edit.xlsx', sheet_name='Sheet 3')
```

```python
top_crimes_df
```

```python
# Function to format population
def millions_formatter(x, pos):
    return f'{int(x/1000000)}M'
```

```python
## Scatterplot of Top Ten States Murder Rate by Population
plt.figure(figsize=(10,8))
sns.scatterplot(data=top_crimes_df, x='murder', y='population', hue='state', marker="$\circ$", ec='face', s=100)
plt.title("Scatterplot of Top Ten States Murder Rate by Population")
plt.xlabel('Murder Rate')
plt.ylabel('Population in Millions')
plt.gca().yaxis.set_major_formatter(FuncFormatter(millions_formatter))
plt.show()
```

```python
## Bubble graph of Population vs Rape Rates: Bubble Size Indicates Murder Rate

plt.figure(figsize=(12,8))
sns.scatterplot(data=top_crimes_df, x='population', y='forcible_rape', marker="$\circ$", size='murder', hue='state', sizes=(50,200))
plt.title("Bubble Graph of Population vs Rape Rates: Bubble Size Indicates Murder Rate")
plt.xlabel("Population in Millions")
plt.ylabel("Forcible Rape Rate")
plt.gca().xaxis.set_major_formatter(FuncFormatter(millions_formatter))
plt.show()
```

```python
# Density Plot of Population
sns.kdeplot(data=top_crimes_df, x='population', fill=True)
plt.title('Density Plot of Population')
plt.xlabel('Population in Milions')
plt.ylabel('Density')
plt.gca().xaxis.set_major_formatter(FuncFormatter(millions_formatter))
plt.show()
```
